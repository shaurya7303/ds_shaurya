```python
td["Timestamp IST"] = pd.to_datetime(
    td["Timestamp IST"],
    format="%d-%m-%Y %H:%M",
    errors="coerce"
)

td["date"] = td["Timestamp IST"].dt.date
```

we extract the date from the timestamp IST for merging


```python
sent["date"] = pd.to_datetime(sent["date"], errors="coerce").dt.date
sent["classification"] = sent["classification"].str.strip().str.title()
```

```python
td = td.rename(columns={
    "Account": "account",
    "Coin": "coin",
    "Execution Price": "execution_price",
    "Size Tokens": "size_tokens",
    "Size USD": "size_usd",
    "Side": "side",
    "Closed PnL": "closed_pnl",
    "Direction": "direction"
})
```

```python
td
```

## Data Cleaning


```python
td = td.dropna(subset=[
    "date", "closed_pnl", "size_usd"
])

sent = sent.dropna(subset=[
    "date", "classification"
])
```

Merging both the dataset

```python
df = td.merge(
    sent[["date", "classification"]],
    on="date",
    how="inner"
)

print("Merged shape:", df.shape)
df.head()
```

FEATURE Engineering

```python
df["is_profitable"] = df["closed_pnl"] > 0
```

```python
df["abs_size_usd"] = df["size_usd"].abs()
df["side_encoded"] = df["side"].map({"BUY": 1, "SELL": -1})
df["pnl_per_usd"] = df["closed_pnl"] / df["abs_size_usd"]
```

we drive the profit indicator , absolute exposure , buy or sell encoding and normalized Pnl

```python
sentiment_summary = df.groupby("classification").agg(
    total_trades=("account", "count"),
    unique_traders=("account", "nunique"),
    avg_pnl=("closed_pnl", "mean"),
    median_pnl=("closed_pnl", "median"),
    win_rate=("is_profitable", "mean"),
    avg_trade_size_usd=("abs_size_usd", "mean")
).reset_index()

sentiment_summary
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(
    data=df,
    x="classification",
    y="closed_pnl"
)
plt.title("PnL distribution across the market sentiment")
plt.xlabel("market Sentiment")
plt.ylabel("closed PnL")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(
    data=df,
    x="classification",
    y="abs_size_usd"
)
plt.title("Trade size distribution by sentiment , Behavior vs Sentiment")
plt.xlabel("market sentiment")
plt.ylabel("trade Size ")
plt.xticks(rotation=30)
plt.show()
```

```python
trader = df.groupby(
    ["account", "classification"]
).agg(
    trades=("closed_pnl", "count"),
    total_pnl=("closed_pnl", "sum"),
    avg_pnl=("closed_pnl", "mean"),
    win_rate=("is_profitable", "mean"),
    avg_trade_size=("abs_size_usd", "mean")
).reset_index()

trader.head()
```

```python
consistent_traders = trader.groupby("account").agg(
    total_pnl=("total_pnl", "sum"),
    avg_win_rate=("win_rate", "mean"),
    total_trades=("trades", "sum")
).reset_index()

consistent_traders = consistent_traders[
    (consistent_traders["total_pnl"] > 0) &
    (consistent_traders["avg_win_rate"] > 0.55)
]

consistent_traders.sort_values("total_pnl", ascending=False).head(10)
```

```python
side_sentiment = df.groupby(
    ["classification", "side"]
).agg(
    trades=("closed_pnl", "count"),
    avg_pnl=("closed_pnl", "mean"),
    win_rate=("is_profitable", "mean")
).reset_index()

side_sentiment
```

```python
df.to_csv("processed_trades.csv", index=False)
sentiment_summary.to_csv("sentiment_summary.csv", index=False)
trader.to_csv("trader_performance.csv", index=False)
print("all output save successfully")
```
